```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
fs = 1000  # Sampling frequency in Hz
t = np.arange(0, 1, 1/fs)  # Time vector from 0 to 1 second
f_carrier = 8  # Theta frequency in Hz
f_carried = 60  # Gamma frequency in Hz
A_carrier = 1
A_carried = 1  # Modulation depth

# Nested frequency signal
carrier_component = -np.cos(2 * np.pi * f_carrier * t)
carried_component = np.cos(2 * np.pi * f_carried * t)
nested_signal = A_carried * carried_component + A_carrier * A_carried * carrier_component * carried_component

# Plotting the nested signal
plt.figure(figsize=(10, 6))
plt.plot(t, nested_signal, label='Nested Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Nested Frequency Signal with Phase-Amplitude Coupling')
plt.legend()
#plt.grid(True)
plt.show()
```
