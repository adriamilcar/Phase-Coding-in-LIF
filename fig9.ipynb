# Cortex

## Neuron params

```python
# Default params (taken from Teeter et al Nature Comms 2018)
R_m = 177 * 1e6                                     # 142 MΩ
V_th = 28.3 * 1e-3                                    # 15 mV
tau_m = 18.9 * 1e-3                                   # 24 ms  

# Other params
I_osc = 40 * 1e-12                                  # 40 pA
range_frac = 0.6                                    # fraction of range to use
tau_s = 50 * 1e-3                                  # 200 ms of decay time constant of signal
f_lims = [1, 50]
eta_lims = [1e-10, 0.3]
```

```python
M = 10
res = 200
fs = np.linspace(f_lims[0], f_lims[1], res)
etas = np.linspace(eta_lims[0], eta_lims[1], res)
mis_theo_ctx = np.zeros((res,res))
for i, eta in enumerate(etas):
    print(i)
    for j, f in enumerate(fs):
        means, variances, Is_range = get_distr(R_m, V_th, eta, tau_m, I_osc, f, M, range_frac, N=1)
        mis_theo_ctx[i, j] = approx_mi(means, variances)
```

```python
info_rate_ctx_norm = get_R(mis_theo_ctx, fs, tau_s, corr=True, norm=True)
```

# Olfactory bulb

```python
# Default params (taken from Margrie and Schaefer 2003 J. of Physiology)
R_m = 100 * 1e6                      # 100 MΩ
V_th = 15 * 1e-3                     # 15 mV
tau_m = 10 * 1e-3                    # 10 ms  
I_osc = 35.4 * 1e-12                 # to create a subthreshold oscillation of 3.45 mV of amplitude for a 3.7 Hz oscillation
tau_s = 250 * 1e-3                   # 250 ms of decay time constant of signal

# Other params                              
range_frac = 0.6                                    # fraction of range to use
f_lims = [1, 50]
eta_lims = [1e-10, 0.3]
```

```python
M = 10
res = 200
fs = np.linspace(f_lims[0], f_lims[1], res)
etas = np.linspace(eta_lims[0], eta_lims[1], res)
mis_theo_ob = np.zeros((res,res))
for i, eta in enumerate(etas):
    print(i)
    for j, f in enumerate(fs):
        means, variances, Is_range = get_distr(R_m, V_th, eta, tau_m, I_osc, f, M, range_frac, N=1)
        mis_theo_ob[i, j] = approx_mi(means, variances)
```

```python
info_rate_ob_norm = get_R(mis_theo_ob, fs, tau_s, corr=True, norm=True)
```

# Plots

```python
plt.figure(figsize=(12.5, 5))

# Apply Gaussian smoothing to the data
sigma = 3  # Standard deviation for Gaussian kernel, adjust as needed
info_rate_ctx_norm_smooth = gaussian_filter(info_rate_ctx_norm, sigma=1e-10)
info_rate_ob_norm_smooth = gaussian_filter(info_rate_ob_norm, sigma=1e-10)

# Determine the common color scale range
vmin = 0
vmax = 1

# Define contour levels using a power law
num_levels = 12
power = 2  # Adjust power to control spacing between levels
contour_levels = np.linspace(0, 1, num=num_levels)**power * (vmax - vmin) + vmin


# Simulations plot
plt.subplot(121)
plt.title('Pyramidal cell (V1)', fontsize=24, pad=10)
img2 = plt.imshow(info_rate_ctx_norm_smooth, origin='lower', cmap='jet', vmin=vmin, vmax=vmax, interpolation='lanczos')
contours2 = plt.contour(info_rate_ctx_norm_smooth, levels=contour_levels, colors='black', linewidths=1.5, origin='lower')
plt.clabel(contours2, inline=True, fontsize=11.5, fmt='%1.2f')  # Adding labels to contours
plt.xticks(np.linspace(0, res-1, 6), np.linspace(f_lims[0], f_lims[1], 6).astype(int), fontsize=20)
plt.yticks(np.linspace(0, res-1, 6), np.linspace(eta_lims[0], eta_lims[1], 6).round(2), fontsize=20)
plt.xlabel('$f$ (Hz)', fontsize=22)
plt.ylabel('$\eta$ (a.u.)', fontsize=22)
cbar = plt.colorbar(img2, label=r'$r_{norm.}$', shrink=.5)
cbar.set_label(r'$r_{norm.}$', fontsize=20)
cbar.ax.tick_params(labelsize=18)  # Increase colorbar tick fontsize

# Theory plot
plt.subplot(122)
plt.title('Mitral cell (OB)', fontsize=24, pad=10)
img2 = plt.imshow(info_rate_ob_norm_smooth, origin='lower', cmap='jet', vmin=vmin, vmax=vmax, interpolation='lanczos')
contours2 = plt.contour(info_rate_ob_norm_smooth, levels=contour_levels, colors='black', linewidths=1.5, origin='lower')
plt.clabel(contours2, inline=True, fontsize=11.5, fmt='%1.2f')  # Adding labels to contours
plt.xticks(np.linspace(0, res-1, 6), np.linspace(f_lims[0], f_lims[1], 6).astype(int), fontsize=20)
plt.yticks(np.linspace(0, res-1, 6), np.linspace(eta_lims[0], eta_lims[1], 6).round(2), fontsize=20)
plt.xlabel('$f$ (Hz)', fontsize=22)
plt.ylabel('$\eta$ (a.u.)', fontsize=22)
cbar = plt.colorbar(img2, label=r'$r_{norm.}$', shrink=.5)
cbar.set_label(r'$r_{norm.}$', fontsize=20)
cbar.ax.tick_params(labelsize=18)  # Increase colorbar tick fontsize


plt.tight_layout()

plt.savefig('plots/fig9/fig9_R.svg', dpi=600)

plt.show()
```
